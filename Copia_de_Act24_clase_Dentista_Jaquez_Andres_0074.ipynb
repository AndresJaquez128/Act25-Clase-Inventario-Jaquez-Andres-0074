```python
# Hecho por Jaquez Andres 0074

# ---------------------------------------------------------
# DEFINICIÓN DE LA CLASE: Inventario (El Plano)
# ---------------------------------------------------------
class Inventario:
    def __init__(self, producto, cantidad, ubicación):
        # --- Atributos: Los Datos (Sustantivos) ---
        self.producto = producto
        self.cantidad = cantidad
        self.ubicación = ubicación

    # --- Funciones: Las Acciones (Verbos) ---
    def leerdatos(self):
        """
        Solicita al usuario los datos del producto por consola.
        """
        print(f"\n📥 --- Capturando nuevos datos para '{self.producto}' ---")
        try:
            self.producto = input("   Escribe el nombre del producto: ")
            self.cantidad = int(input(f"   Escribe la cantidad (actual: {self.cantidad}): "))
            self.ubicación = input(f"   Escribe la ubicación (actual: {self.ubicación}): ")
            print("   ✅ Datos de producto actualizados correctamente.")
        except ValueError:
            print("   ⚠️ ERROR: La cantidad debe ser un número entero. Los cambios no se guardaron.")

    def mostrardatos(self):
        """
        Muestra la ficha técnica del producto en consola.
        """
        print("\n📄 ==============================")
        print("   FICHA TÉCNICA DEL PRODUCTO   ")
        print("   ==============================")
        print(f"   Producto:    {self.producto}")
        print(f"   Cantidad:    {self.cantidad}")
        print(f"   Ubicación:   {self.ubicación}")
        print("   ==============================")
# ---------------------------------------------------------
# CREACIÓN Y USO DE OBJETOS (Ejecución y Salida de Consola)
# ---------------------------------------------------------
# He incluido un menú interactivo para que puedas probar ambos objetos y sus funciones las veces que quieras.

def menu():
    # 1. Instanciación: Crear objetos iniciales con datos de ejemplo
    inventario1 = Inventario("Sillas de Oficina", 15, "Pasillo B-3")
    inventario2 = Inventario("Mesas de Conferencia", 10, "Área Común")

    while True:
        print("\n" + "="*40)
        print("      SISTEMA DE GESTIÓN DE INVENTARIO")
        print("="*40)
        print(f"1. Gestionar Objeto 1: {inventario1.producto}")
        print(f"2. Gestionar Objeto 2: {inventario2.producto}")
        print("3. Salir del programa")

        opcion_principal = input("\nSeleccione un objeto (1-3): ")

        if opcion_principal == "3":
            print("\n👋 Programa terminado. Hecho por Jaquez Andres 0074.")
            break

        # Selección del objeto según la opción elegida
        if opcion_principal == "1":
            obj = inventario1
        elif opcion_principal == "2":
            obj = inventario2
        else:
            print("⚠️ Opción no válida. Intenta de nuevo.")
            continue

        # Submenú de acciones (Métodos) para el objeto seleccionado
        while True:
            print(f"\n--- Panel de Gestión: '{obj.producto}' ---")
            print(" a) leerdatos() - Actualizar información")
            print(" b) mostrardatos() - Ver ficha actual")
            print(" c) Volver al menú principal")

            accion = input("¿Qué acción desea realizar? (a, b o c): ").lower()

            # Bloque de ejecución: Llamada a los métodos del objeto
            if accion == "a":
                # Llama a la función leerdatos() del objeto seleccionado
                obj.leerdatos()
                input("\nPresiona Enter para volver...")
            elif accion == "b":
                # Llama a la función mostrardatos() del objeto seleccionado
                obj.mostrardatos()
                input("\nPresiona Enter para volver...")
            elif accion == "c":
                # Rompe el bucle interno para volver al menú principal
                break
            else:
                print("⚠️ Acción no reconocida. Intenta de nuevo.")

# Ejecutar el menú interactivo al correr el script
if __name__ == "__main__":
    menu()
```
